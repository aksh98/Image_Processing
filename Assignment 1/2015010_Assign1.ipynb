```
# -----------------------------------------
# Backward Nearest Neighbour Interpolation
# -----------------------------------------


def interpolation(inpu,outut,int_fac):
  for i in range(0,len(outut)):
    for j in range(0,len(outut[0])):
      
      tempx = (i)/int_fac 
      tempy = (j)/int_fac 
      
      if(tempx-int(tempx) > int(tempx+1.0)-tempx):
        if(tempy - int(tempy) > int(tempy+1.0)-tempy):
          outut[i][j] = inpu[int(tempx+1)][int(tempy+1)]
        else:
          outut[i][j] = inpu[int(tempx+1)][int(tempy)]
      else:
        if(tempy - int(tempy) > int(tempy+1.0)-tempy):
          outut[i][j] = inpu[int(tempx)][int(tempy+1)]
        else:
          outut[i][j] = inpu[int(tempx)][int(tempy)]
#       print(outut)
     
  return outut
```

```
# one extra index to ignore the zeroth index

# inputi = [[0,0,0],[0,6,7],[0,8,9]]
output = [[0 for i in range(4)] for i in range(4)]
inputi = [[6,7],[8,9]]
int_fac = len(output)/len(inputi)
interpolation(inputi,output,int_fac)
```
